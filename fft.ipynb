### Perioddiagram

```python
import numpy as np
import pylab as plt
np.random.seed(1234)
```

```python
fs = 10e3
N = 1e5
amp = 2*np.sqrt(2)
freq = 1234.0
noise_power = 0.001 * fs / 2
time = np.arange(N) / fs
x = amp*np.sin(2*np.pi*freq*time)
x += np.random.normal(scale=np.sqrt(noise_power), size=time.shape)
```

```python
# f, Pxx_den = signal.periodogram(x, fs)
dft = np.fft.fft(x*np.hamming(len(x)))
dft = np.fft.fftshift(dft)
dft = np.abs(dft)**2/(len(x)/1e4)
xft = np.fft.fftfreq(len(x),d=1./fs)
xft = np.fft.fftshift(xft)
plt.semilogy(xft, dft)
# plt.ylim([1e-7, 1e2])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
